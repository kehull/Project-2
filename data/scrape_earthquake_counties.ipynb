```python
import requests
import pandas as pd
```

```python
df = "California_EarthQuake.csv"
quake_df = pd.read_csv(df)
quake_loc = quake_df[['place']]
quake_loc
```

```python
i=0
for index,row in quake_loc.iterrows():
    city = row[0]
    quake_loc.at[i,'citystate'] = city.split('of ')[1]
    i=i+1
quake_loc
```

```python
citystate_list = quake_loc['citystate'].to_list()
i=0
for city in citystate_list:
    if city.find('Is.') != -1:
        citystate_list.remove(city)
        city2 = city.split(" ")
        city2 = city2[0:2]
        city2 = ' '.join(map(str, city2))
        city2 = f'{city2}, CA'
        citystate_list.insert(i,city2)
        print(f'Rewrote the name of {city} to {city2}')
        i=i+1
    elif city.find('Geysers') != -1:
        citystate_list.remove(city)
        city2 = city.split(",")[0]
        citystate_list.insert(i,city2)
        print(f'Rewrote the name of {city} to {city2}')
        i = i+1
    else:
        i=i+1
print(citystate_list)
```

```python
i=0
for city in citystate_list:
    if city.find(',') != -1:
        try:
            #build search query
            search_url = "http://uscounties.com/zipcodes/search.pl?query="
            search_query = city
            updated_query = search_query.replace(" ","+")
            updated_query = updated_query.replace(",","%2C")
            url = f'{search_url}{updated_query}&stpos=0&stype=AND'
            r = requests.get(url)
            df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
            df = df_list[2]
            county_df = df[2]
            county = county_df[1].title()
            quake_loc.at[i,'county'] = county
            print(county)
            i=i+1
            pass
        
        except:
            #build search query
            search_url = "https://en.wikipedia.org/wiki/"
            search_query = city
            url = f'{search_url}{search_query}'
            r = requests.get(url)
            df_list = pd.read_html(r.text) # this parses all the tables in webpages to a list
            df = df_list[0]
            county_row = df[df.eq("County").any(1)]
            thelist = county_row.values.tolist()[-1]
            county = thelist[-1].title()
            quake_loc.at[i,'county'] = county
            print(county)
            i=i+1
    elif city.find('Geysers') != -1:
        county = 'Sonoma'
        quake_loc.at[i,'county'] = county
        print(county)
        i=i+1
    else:
        county = f'{city} is not a city'
        quake_loc.at[i,'county'] = 'n/a'
        print(county)
        i = i+1
print('Scraping complete.')
```
